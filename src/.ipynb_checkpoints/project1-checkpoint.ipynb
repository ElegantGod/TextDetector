```python
#!usr/bin/env python
# Title: project1_finished
# description: simulate the lateral inhibitory of human eyes.
# author: Siyu Zhu
# Create Date: Oct 2, 2014
# Modified Date: OCt 2, 2014
# usage:
# python version: 2.7
import scipy 
import math
import numpy as np
from scipy.misc import imread, imsave, imshow
from scipy.signal import convolve2d
import matplotlib.pyplot as plt

def imgLoad(filename):
    img = imread(filename, flatten = True)
    img = img - np.min(img)
    img = img / np.max(img)
    return img

def lateralInhabitoryFilter(sigma1, sigma2, rat):
    pi = 3.14
    fsize = 11
    x0 = y0 = int(math.floor(fsize/2))

    rat = 1.5
    sigma1 = 3.0
    sigma2 = 15.0
    excitatory = np.empty((fsize, fsize))
    inhibitory = np.empty((fsize, fsize))
    for i in range(fsize):
        for j in range(fsize):
            excitatory[i, j] = 1/(sigma1*(2*pi)**(0.5)) * math.exp(-((i - x0)**2+(j-y0)**2)/(2*sigma1**2))
            inhibitory[i, j] = 1/(sigma2*(2*pi)**(0.5)) * math.exp(-((i - x0)**2+(j-y0)**2)/(2*sigma2**2))
    fil = rat * excitatory - inhibitory
    return fil

# Too slow, just for showcase
def filterResponse0(fil, img):
    fsize = fil.shape[0]
    img1 = np.empty((img.shape[0] - fsize, img.shape[1] - fsize))
    for p in range(img1.shape[0]):
        for q in range(img1.shape[1]):
            s = 0
            for i in range(fsize):
                for j in range(fsize):
                    s += img[p+i, q+j]*li[i, j]
            img1[p, q] = s
    return img1

def filterResponse1(fil, img):
    fsize = fil.shape[0]
    img1 = np.empty((img.shape[0] - fsize, img.shape[1] - fsize))
    for p in range(img1.shape[0]):
        for q in range(img1.shape[1]):
            img1[p, q] = np.sum(img[p:p+fsize, q:q+fsize]*fil)
    return img1

def filterResponse2(fil, img):
    img1 = convolve2d(img, fil, 'valid')
    return img1


def imgSave(filename, img):
    imsave(filename, img)

img = imgLoad('Lenna.png')
fil = lateralInhabitoryFilter(3.0, 15.0, 1.5)
new_img = filterResponse2(fil, img)
# plt.imshow(new_img)
imgSave('output.png', new_img)
```
