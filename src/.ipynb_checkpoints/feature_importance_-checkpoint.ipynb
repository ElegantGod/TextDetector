```python
import pickle
import matplotlib.pyplot as plt
```

```python
clfname = '../clf/icdar2013wordadaboost'
clf = pickle.load(open(clfname))
fi = clf.feature_importances_
fi = numpy.reshape(fi, [3, 3, 1000])
```

```python
def grid_plot(fi):
    f1 = numpy.sum(fi, axis = 2)
    fig = plt.figure()
    plt.clf()
    ax = fig.add_subplot(111)
    ax.set_aspect(1)
    res = ax.imshow(f1, cmap=plt.cm.hot, interpolation='nearest')
    cb = fig.colorbar(res)
    plt.title('feature importance across grids')
```

```python
def codebook_plot(fi):
    f2 = numpy.sum(numpy.sum(fi, axis = 0), axis = 0)
    plt.plot(numpy.sort(f2))
    plt.title('feature importance across codebooks (sorted)')
```

```python
grid_plot(fi)
```

```python
codebook_plot(fi)
```
